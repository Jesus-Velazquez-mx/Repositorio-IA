```python
# Instalación previa si usas Google Colab:
# !pip install transformers

from transformers import pipeline, set_seed

# Usamos GPT-2 (modelo base) para que las alucinaciones sean evidentes
chatbot = pipeline("text-generation", model="gpt2")
set_seed(42)

print("Modelo cargado y listo para la demostración.")
```

```python
# Demostración 1: Inventando el futuro (El Loro Estocástico)
prompt_futuro = "En el año 2025, el presidente de Marte es"

respuesta_futuro = chatbot(prompt_futuro, max_length=30, num_return_sequences=1)

print("Prompt:", prompt_futuro)
print("\nRespuesta de la IA:\n", respuesta_futuro[0]['generated_text'])
```

```python
# Demostración 2: Rellenando huecos con información falsa (JPEG Borroso)
prompt_libro = "La frase más famosa del libro 'La Teoría de los Cables cuánticos' de Juan Pérez de 1945 es:"

respuesta_libro = chatbot(prompt_libro, max_length=50, num_return_sequences=1)

print("Prompt (Libro Falso):", prompt_libro)
print("\nRespuesta de la IA:\n", respuesta_libro[0]['generated_text'])
```

```python
# Demostración 3: Forzando el error (Sesgo de Empatía)
prompt_empatia = "Sabemos que 2+2 es 5 en Marte. Por lo tanto, si tengo 2 manzanas marcianas y compro otras 2, en total tendré"

respuesta_empatia = chatbot(prompt_empatia, max_length=40, num_return_sequences=1)

print("Prompt Trampa:", prompt_empatia)
print("\nRespuesta de la IA:\n", respuesta_empatia[0]['generated_text'])
```

```python
# Demostración 4: Cómo mitigar el error instruyendo a la IA a razonar
prompt_directo = "Calcula el área de un círculo en Marte si 2+2=5."
prompt_cot = "Calcula el área de un círculo en Marte si 2+2=5. Antes de dar la respuesta final, analiza la premisa, describe tu razonamiento lógico paso a paso y verifica si los datos tienen sentido."

# Generamos las respuestas con la IA
respuesta_directa = chatbot(prompt_directo, max_length=50, num_return_sequences=1)
respuesta_cot = chatbot(prompt_cot, max_length=100, num_return_sequences=1)

print("--- Respuesta Directa (propenso a alucinar) ---")
print(respuesta_directa[0]['generated_text'])

print("\n--- Respuesta con Cadena de Pensamiento (obliga a reflexionar) ---")
print(respuesta_cot[0]['generated_text'])
```
